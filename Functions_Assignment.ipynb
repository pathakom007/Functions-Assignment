1. What is the difference between a function and a method in Python ?

 - A function in Python is a block of reusable code that performs a specific task.

 - It is defined using the def keyword.

 - Functions can exist independently (not tied to any object/class).

```python
def add(a, b):
    return a + b

print(add(5, 3))  # Output: 8
```

 - A method is also a function, but it is associated with an object (i.e., it belongs to a class).

 - It is defined inside a class and usually takes self as its first parameter (to access the instance of the class).

 - Methods are called on objects.

```python
class Calculator:
    def add(self, a, b):   # This is a method
        return a + b

calc = Calculator()
print(calc.add(5, 3))  # Output: 8
```

2. Explain the concept of function arguments and parameters in Python ?

 - Parameters are placeholders/variables defined inside the function definition.

```python
def show(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show(1, 2, 3, name="Om", age=21)
# Args: (1, 2, 3)
# Kwargs: {'name': 'Om', 'age': 21}
```

5. What are iterators in Python and how do they differ from iterables ?

1. Iterable :

      - An iterable is any Python object that can be looped over (like in a for loop).

      - It has an __iter__() method that returns an iterator.

      - Examples: lists, tuples, strings, dictionaries, sets, ranges.

```python
my_list = [1, 2, 3]
for x in my_list:   # list is iterable
    print(x)
```

2. Iterator :

      - An iterator is an object that produces values one at a time using the __next__() method.

      - It is created from an iterable using the iter() function.

      - When there are no more items, it raises a StopIteration exception.

```python
my_list = [1, 2, 3]

it = iter(my_list)   # create iterator
print(next(it))      # 1
print(next(it))      # 2
print(next(it))      # 3
# print(next(it))    # StopIteration error
```

6. Explain the concept of generators in Python and how they are defined ?

1. Concept

     - A generator is a special type of function in Python that produces a sequence of values one at a time.

     - Unlike normal functions that use return, generators use yield.

     - They don’t return all values at once → they generate values lazily (on demand).

     - Every generator is an iterator, but you don’t need to write the __iter__ () and __next__() methods yourself.

2. How to Define a Generator

      - Use the yield keyword instead of return.

      - When a generator function is called, it does not run immediately → it returns a generator object.

      - You can get values using next() or in a for loop.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i   # pauses here and returns i
        i += 1

gen = count_up_to(5)

print(next(gen))  # 1
print(next(gen))  # 2
print(next(gen))  # 3
```

7. What are the advantages of using generators over regular functions ?

Advantages of Generators over Regular Functions :

1. Memory Efficiency
2. Lazy Evaluation (On-demand computation)
3. Infinite Sequences
4. Better Performance for Large Data
5. Cleaner Code for Iterators

```python
def squares_list(n):
    result = []
    for i in range(n):                            # Regular Function (returns list → uses more memory)
        result.append(i * i)
    return result

# Using the function
nums = squares_list(5)
print(nums)   # [0, 1, 4, 9, 16]
```

```python
def squares_generator(n):
    for i in range(n):
        yield i * i                              # Generator Function (yields one value at a time → memory efficient)

# Using the generator
nums = squares_generator(5)
for num in nums:
    print(num)
```

9. Explain the purpose and usage of the `map()` function in Python ?

   - The map() function applies a given function to each item of an iterable (like a list, tuple, etc.) and returns a map object (an iterator) with the results.

Example : Using map() with a lambda function

```python
nums = [1, 2, 3, 4, 5]
squares = map(lambda x: x**2, nums)
print(list(squares))   # [1, 4, 9, 16, 25]
```

Usage:

1. Transforms data efficiently – You don’t need a loop to apply a function to each element.

2. Works with multiple iterables – Can combine lists element-wise.

3. Memory-efficient – Returns a generator-like map object, not a list, until you convert it.

4. Often used with lambda for short, inline operations.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?

1. map()

  - Purpose: Apply a function to each element of an iterable and return a new iterable with the results.

  - Returns: A map object (iterator)

```python
nums = [1, 2, 3, 4]
squares = map(lambda x: x**2, nums)
print(list(squares))   # [1, 4, 9, 16]
```

2. filter()

  - Purpose: Select elements from an iterable that satisfy a condition.

  - Returns: A filter object (iterator) containing only elements where the function returns True.

```python
nums = [1, 2, 3, 4, 5]
even_nums = filter(lambda x: x % 2 == 0, nums)
print(list(even_nums))   # [2, 4]
```

3. reduce()

  - Purpose: Apply a function cumulatively to elements of an iterable, reducing it to a single value.

  - Note: reduce() is in the functools module.

```python
from functools import reduce

nums = [1, 2, 3, 4]
sum_nums = reduce(lambda x, y: x + y, nums)
print(sum_nums)   # 10
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.



```python
def sum_of_evens(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:  # Check if the number is even
            total += num
    return total

# Example usage
nums = [1, 2, 3, 4, 5, 6]
print(sum_of_evens(nums))  # Output: 12
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n       # Number of terms
        self.i = 0       # Current index
        self.a, self.b = 0, 1  # First two Fibonacci numbers

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        if self.i == 0:
            self.i += 1
            return 0
        elif self.i == 1:
            self.i += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.i += 1
            return self.b

# Example usage
for num in Fibonacci(10):
    print(num, end=' ')
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.



```python
def powers_of_two(n):
    i = 0
    while i <= n:
        yield 2 ** i
        i += 1

# Example usage
for value in powers_of_two(5):
    print(value, end=' ')

```

7. Implement a generator function that reads a file line by line and yields each line as a string.



```python
# Create a sample file called 'example.txt'
with open('example.txt', 'w') as f:
    f.write("Hello World\n")
    f.write("This is line 2\n")
    f.write("This is line 3\n")

def read_file(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.rstrip('\n')  # Remove newline characters

for line in read_file('example.txt'):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius = [0, 20, 37, 100]

# Function to convert Celsius to Fahrenheit
def c_to_f(c):
    return (c * 9/5) + 32

# Using map to convert each temperature
fahrenheit = list(map(c_to_f, celsius))

print(fahrenheit)
```
